```python
import numpy as np 
from matplotlib import pyplot as plt 

FILENAME = "./results/abcnn_title.txt"
```

```python
with open(FILENAME,'r',encoding="utf-8") as fin:
    lines = fin.readlines()[:-2]
    lossLst = []
    accLst = []
    for line in lines:
        if "loss" in line:
            lossLst.append(float(line.split("lossï¼š")[-1].strip()))
        elif "accu" in line:
            accLst.append(float(line.split("accu:")[-1].strip()))
        else:
            continue
assert len(lossLst) == len(accLst)
print(len(lossLst))
print(lossLst)
print(len(accLst))
print(accLst)
```

```python
X = np.arange(len(lossLst))
Y1 = np.array(lossLst)
Y2 = np.array(accLst)
print(X)
fig,ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(X,Y1,label="Loss")
ax1.set_ylabel("Loss")
ax1.set_xlabel("Epoch")
ax2.plot(X,Y2,label="Accu",color="orange")
ax2.set_ylabel("Accu")
plt.title("abcnn lr=1e-3 bsz=32")
fig.legend(loc="upper left", bbox_to_anchor=(1, 0), bbox_transform=ax1.transAxes)
plt.savefig("./abcnn.png", bbox_inches='tight')
plt.show()
# print(Y1)
# print(Y2)
```
